# Credit Default Prediction — 신용 불이행 예측 (KOAI 2026)

고객의 소득·나이·부채비율 등 표 feature 로 **채무 불이행**(default)을 예측한다. `data/train.csv`(라벨 `default` 포함)로 학습하고 `data/test.csv`(라벨 없음)를 예측해
`submission.csv`(`id, default`=**확률**)를 만든다. 평가: **ROC-AUC**(held-out).

```python
import pandas as pd, numpy as np
train = pd.read_csv("data/train.csv")
test  = pd.read_csv("data/test.csv")
X = train.drop(columns=["id", "default"]); y = train["default"]
Xtest = test.drop(columns=["id"])
print(train.shape, "| default rate", round(y.mean(), 3))
```

```python
# 베이스라인: 모두에게 학습셋 평균확률(=변별력 없음) → AUC ≈ 0.5
pd.DataFrame({"id": test["id"], "default": float(y.mean())}).to_csv("submission.csv", index=False)
print("saved submission.csv", len(test))
```

## 제출 파일 모으기
아래 셀을 실행하면 제출 파일이 **최상위(`/content`)로 복사**되어 왼쪽 파일 탐색기에 바로 보입니다.
그 파일을 내려받아 연습 사이트 **Submissions** 탭에 올리면 채점됩니다.

```python
# === 제출 파일을 /content 로 모으기 (마지막에 실행) ===
import os, glob, shutil
TARGETS = ['submission.csv']
OUT = "/content" if os.path.isdir("/content") else os.getcwd()
found = []
for name in TARGETS:
    hits = [name] if os.path.exists(name) else glob.glob(f"**/{name}", recursive=True)
    if not hits:
        print("아직 없음(해당 셀을 먼저 실행하세요):", name); continue
    dst = os.path.join(OUT, os.path.basename(hits[0]))
    if os.path.abspath(hits[0]) != os.path.abspath(dst):
        shutil.copy2(hits[0], dst)
    found.append(dst)
print("제출 파일 저장 위치(파일 탐색기 최상위):", found)
```
