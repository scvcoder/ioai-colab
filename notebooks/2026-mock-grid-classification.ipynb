# Grid-Collage Image Classification — 궁격도 분류 (APOAI 2026 Mock)

이미지가 **궁격도(grid-collage, 여러 이미지를 격자로 콜라주한 것)**인지 아닌지 이진분류한다. **비지도** 과제 —
`train.csv` 라벨은 목표(격자 여부)가 아니라 상품 카테고리(여성복/뷰티)라 사실상 라벨이 없다. 비지도 CV 나
수동 라벨링으로 접근한다. `data/test/`(+`test_list.csv`)를 예측해 `submission.csv`(filename,label). 채점: accuracy.
(사이트 채점용 정답은 손으로 라벨한 held-out 48장.)

```python
import pandas as pd, numpy as np, os
from PIL import Image
train = pd.read_csv("data/train.csv")            # filename (라벨 없음)
test_list = pd.read_csv("data/test_list.csv")["filename"].tolist()
print("train", len(train), "| test", len(test_list))
```

```python
# 베이스라인: 전부 비격자(0) — accuracy ≈ 0.5
pd.DataFrame({"filename": test_list, "label": [0]*len(test_list)}).to_csv("submission.csv", index=False)
print("saved submission.csv", len(test_list))
```

## 제출 파일 모으기
아래 셀을 실행하면 제출 파일이 **최상위(`/content`)로 복사**되어 왼쪽 파일 탐색기에 바로 보입니다.
그 파일을 내려받아 연습 사이트 **Submissions** 탭에 올리면 채점됩니다.

```python
# === 제출 파일을 /content 로 모으기 (마지막에 실행) ===
import os, glob, shutil
TARGETS = ['submission.csv']
OUT = "/content" if os.path.isdir("/content") else os.getcwd()
found = []
for name in TARGETS:
    hits = [name] if os.path.exists(name) else glob.glob(f"**/{name}", recursive=True)
    if not hits:
        print("아직 없음(해당 셀을 먼저 실행하세요):", name); continue
    dst = os.path.join(OUT, os.path.basename(hits[0]))
    if os.path.abspath(hits[0]) != os.path.abspath(dst):
        shutil.copy2(hits[0], dst)
    found.append(dst)
print("제출 파일 저장 위치(파일 탐색기 최상위):", found)
```
