# JOAI 2026 — Lever Decoding (neural regression)

> **Japan Olympiad in AI 2026 — Playground Competition**

Decode the **`lever`** value (≈0–4.3) from a mouse's **cortical brain-region activity** — ~44 signals across left/right hemispheres (AUDp, MOp, SSp, VIS…). Score = **MSE** (lower is better).

The official Kaggle test labels are hidden, so we grade on a **group held-out split** (every 5th `sample_id` — whole trials held out, no within-trial leakage) of `train.csv`. **Submit** `submission.csv` (`id,lever`).

⚠️ **The baseline below is a linear model (Ridge) on the brain signals (MSE ≈ 1.34).** Neural→behaviour decoding is non-linear — switch to a gradient-boosting model and add **time-lagged neural features** (see the model solution) to lower the MSE.

```python
import pandas as pd, numpy as np
df = pd.read_csv('data/train.csv').sort_values('sample_id').reset_index(drop=True)
# GROUP held-out: every 5th sample_id (whole trials held out, no timestep leakage)
sids = sorted(df['sample_id'].unique()); val_set = set(sids[::5])
is_val = df['sample_id'].isin(val_set)
tr, va = df[~is_val], df[is_val]
BRAIN = [c for c in df.columns if c not in ['id','sample_id','mouse_id','lever','day_n','time'] and df[c].dtype != 'O']
print('train', len(tr), 'val', len(va), '| brain-region features:', len(BRAIN))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
m = Ridge(alpha=1.0).fit(tr[BRAIN].fillna(0), tr['lever'])
pred = m.predict(va[BRAIN].fillna(0))
print('held-out MSE:', round(mean_squared_error(va['lever'], pred), 4))
pd.DataFrame({'id': va['id'].to_numpy(), 'lever': pred}).to_csv('submission.csv', index=False)
print('wrote submission.csv', len(va), '(linear baseline — try GBM + lagged neural features)')
```

## 제출 파일 모으기
아래 셀을 실행하면 제출 파일이 **최상위(`/content`)로 복사**되어 왼쪽 파일 탐색기에 바로 보입니다.
그 파일을 내려받아 연습 사이트 **Submissions** 탭에 올리면 채점됩니다.

```python
# === 제출 파일을 /content 로 모으기 (마지막에 실행) ===
import os, glob, shutil
TARGETS = ['submission.csv']
OUT = "/content" if os.path.isdir("/content") else os.getcwd()
found = []
for name in TARGETS:
    hits = [name] if os.path.exists(name) else glob.glob(f"**/{name}", recursive=True)
    if not hits:
        print("아직 없음(해당 셀을 먼저 실행하세요):", name); continue
    dst = os.path.join(OUT, os.path.basename(hits[0]))
    if os.path.abspath(hits[0]) != os.path.abspath(dst):
        shutil.copy2(hits[0], dst)
    found.append(dst)
print("제출 파일 저장 위치(파일 탐색기 최상위):", found)
```
