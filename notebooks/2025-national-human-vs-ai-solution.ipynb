# Human vs AI — 모범답안 (글자 단위 char_wb TF-IDF + LinearSVC)
원대회 reference: 최소 전처리(소문자·공백정리)로 **노이즈 문자**(구두점·유니코드·이모지)를 *특징으로 보존*하고,
**글자 단위 `analyzer='char_wb'` TF-IDF + LinearSVC** 로 분류 — AI 글 특유의 정형화된 문자 패턴을 잡습니다.
채점 = held-out(`ID%5==0`) **accuracy**(높을수록 좋음).

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, f1_score

df = pd.read_csv('data/train.csv'); df['label'] = df['label'].astype(float).astype(int)
df['text'] = df['text'].fillna('').str.lower().str.strip()   # 최소 전처리(노이즈 보존)
is_val = df['ID'] % 5 == 0
tr, va = df[~is_val].reset_index(drop=True), df[is_val].reset_index(drop=True)
```

```
vec = TfidfVectorizer(analyzer='char_wb', ngram_range=(1,4), max_features=100000)
Xtr = vec.fit_transform(tr['text']); Xva = vec.transform(va['text'])
clf = LinearSVC(C=1.0).fit(Xtr, tr['label'])
pred = clf.predict(Xva)
print('held-out accuracy:', round(accuracy_score(va['label'], pred),4),
      '| macro_f1:', round(f1_score(va['label'], pred, average='macro'),4))
pd.DataFrame({'ID': va['ID'], 'label': pred}).to_csv('submission.csv', index=False)
print('wrote submission.csv', len(va))
```

## 제출 파일 모으기
아래 셀을 실행하면 제출 파일이 **최상위(`/content`)로 복사**되어 왼쪽 파일 탐색기에 바로 보입니다.
그 파일을 내려받아 연습 사이트 **Submissions** 탭에 올리면 채점됩니다.

```
# === 제출 파일을 /content 로 모으기 (마지막에 실행) ===
import os, glob, shutil
TARGETS = ['submission.csv']
OUT = "/content" if os.path.isdir("/content") else os.getcwd()
found = []
for name in TARGETS:
    hits = [name] if os.path.exists(name) else glob.glob(f"**/{name}", recursive=True)
    if not hits:
        print("아직 없음(해당 셀을 먼저 실행하세요):", name); continue
    dst = os.path.join(OUT, os.path.basename(hits[0]))
    if os.path.abspath(hits[0]) != os.path.abspath(dst):
        shutil.copy2(hits[0], dst)
    found.append(dst)
print("제출 파일 저장 위치(파일 탐색기 최상위):", found)
```
