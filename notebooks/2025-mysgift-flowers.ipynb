````python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
````

````python
def visualize_xyzrgb(xyzrgb, stride=1, point_size=1, elev=0, azim=0):
    """
    xyzrgb: np.ndarray of shape (N, 6) with columns x,y,z,r,g,b
    stride: take every k-th point to speed up plotting
    point_size: marker size passed to scatter
    elev, azim: camera angles
    """
    pts = xyzrgb[::stride]

    x = pts[:, 0]
    y = pts[:, 1]
    z = pts[:, 2]
    rgb = pts[:, 3:6]

    # Normalize colors to [0,1] if needed
    if rgb.dtype.kind in "iu" or rgb.max() > 1.0:
        rgb = rgb.astype(np.float32) / 255.0

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection="3d")

    ax.scatter(x, z, y, s=point_size, c=rgb, depthshade=False)

    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_zlabel("y")

    # Nice bounds/aspect
    ax.set_xlim(float(x.min()), float(x.max()))
    ax.set_ylim(float(z.min()), float(z.max()))
    ax.set_zlim(float(y.min()), float(y.max() + 1))  # add 1 so the plane is visible
    try:
        ax.set_box_aspect((x.ptp(), y.ptp(), max(1.0, z.ptp())))
    except Exception:
        pass  # older Matplotlib

    # ax.set_axis_off()
    # plt.tight_layout()
    plt.show()
````

## 제출 파일 모으기
아래 셀을 실행하면 제출 파일이 **최상위(`/content`)로 복사**되어 왼쪽 파일 탐색기에 바로 보입니다.
그 파일을 내려받아 연습 사이트 **Submissions** 탭에 올리면 채점됩니다.

````python
# === 제출 파일을 /content 로 모으기 (마지막에 실행) ===
import os, glob, shutil
TARGETS = ['submission.csv', 'submission.zip', 'submission.jsonl', 'submission.json']
OUT = "/content" if os.path.isdir("/content") else os.getcwd()
found = []
for name in TARGETS:
    hits = [name] if os.path.exists(name) else glob.glob(f"**/{name}", recursive=True)
    if not hits:
        print("아직 없음(해당 셀을 먼저 실행하세요):", name); continue
    dst = os.path.join(OUT, os.path.basename(hits[0]))
    if os.path.abspath(hits[0]) != os.path.abspath(dst):
        shutil.copy2(hits[0], dst)
    found.append(dst)
print("제출 파일 저장 위치(파일 탐색기 최상위):", found)
````
